**Task4: Data Cleaning**

```python
#identifying the number of null columns within the dataset
df.isnull().sum()
```

```python
#dropping missing value rows as opposed to replacing 
df.dropna(inplace = True) 
```

```python
#identifying the number of duplicates within the dataset
df.duplicated().sum()
```

```python
#identifying duplicate data
df[df.duplicated()]
```

```python
#dropping duplicate rows
df.drop_duplicates(inplace= True)

```

```python
#confirming deletion of duplicate rows
df[df.duplicated()]
```

```python
#identifying the datatypes of each column to determine if they need amending
df.dtypes
```

```python
#changing datatypes 
df = df.astype({
    'id' : 'object',
    'host_id' : 'object',
    'beds' : 'object'
})
```

### EDA

**Task5: Data Analysis**

**Univariate Analysis**

*These are analyses performed on individual columns.*

```python
#listing the price column of the dataset 
df['price']
```

```python
#identifying outliers in price - boxplots help with identifying outliers
df = df[df['price'] < 1500]

sns.boxplot(data = df, x = 'price')
```

*In the above figure, we can see that most of the data is uniformly spready next to each other - showing that there is no outlier with this constraint*

```python
#creating a price-based histogram
plt.figure(figsize=(8,5))
plt.title('Price Distribution')
sns.histplot(data = df, x = 'price', bins = 100) 
plt.ylabel('Frequency')
plt.show()
```

*After 400 we can see that there are very few transactions, showing that most occur withint he 0 - 400 range*

```python
#creating an availability-based histogram
plt.figure(figsize=(8,5))
plt.title('Availability Distribution')
sns.histplot(data = df, x = 'availability_365') 
plt.ylabel('Frequency')
plt.show()
```

```python
#determining the mean price of each neighbourhood group 
df.groupby('neighbourhood_group')['price'].mean()
```

```python
#determining the mean price, number of reviews, and beds of each neighbourhood group
df.groupby('neighbourhood_group')[['price','number_of_reviews', 'beds']].mean()
```

Feature Engineering

```python
#creating a new column to support further analysis
df['price_per_bed'] = df['price'] / df['beds']

print(df['price_per_bed'])
```

```python
#analysing the different neighbourhoods by the average price per bed instead of price to derive a great level of detail and align for individual travelers
df.groupby('neighbourhood_group')['price_per_bed'].mean()
```

**Bi Variate Analysis** 

*Analysing the relationship between one column with another*

```python
#price dependancy on neighbourhood using barplot
sns.barplot(data = df, x = 'neighbourhood_group', y = 'price', hue = 'room_type')
```

```python
#number of reviews, and the relationship with price using scatterplot
plt.figure(figsize=(8,5))
plt.title('Reviews and Price data based on Beds for Bronx')
sns.scatterplot(data = df[df['neighbourhood_group'] == 'Bronx'], y = 'number_of_reviews', x = 'price', hue = 'beds')
plt.show()
```

```python
#creating a pairplot using price, minimum nights, number of reviews, and availability for a macro-level review
sns.pairplot(data =df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue = 'room_type')
```

```python
#geographical distribution of AirBnB listings
plt.figure(figsize=(10,7))
plt.title('Geographical Distribution of Listings')
sns.scatterplot(data = df, x = 'longitude', y = 'latitude', hue = 'room_type')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
#identifying the correlation of one variable with others
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr
```

```python
#creating a heatmap based on correlation 
plt.figure(figsize=(8, 8))
plt.title('Correlation HeatMap')
sns.heatmap(data = corr, annot= True)
plt.show()
```
